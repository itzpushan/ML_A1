```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
df = pd.read_csv('/content/WineQT.csv')

# Display the first few rows and information about the dataset
df.head()
df.info()
```

```python
# Convert 'quality' values: 1 if greater than 6, otherwise 0
df['quality'] = df['quality'].apply(lambda x: 1 if x > 6 else 0)
df.head()

# Initialize the MinMaxScaler and fit-transform the data excluding the 'quality' column
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df.drop('quality', axis=1))

# Create a new DataFrame with the scaled data and the 'quality' column
scaled_df = pd.DataFrame(scaled_data, columns=df.drop('quality', axis=1).columns)
scaled_df['quality'] = df['quality']

# Display the new DataFrame
scaled_df.head()
```
